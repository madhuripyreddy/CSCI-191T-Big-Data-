```python
import pandas as pd 
df = pd.read_csv("./movies_metadata.csv");
df.head()
```

```python
df['budget']
```

```python
notlist = []
removeList = []
for index,EveryRow in df.iterrows():
    try:
        value = int(EveryRow['budget'])
        if value != 0:
            removeList.append(index)
    except ValueError or TypeError:
        notlist.append(index)
        removeList.append(index)
print(notlist)
print(removeList)
```

```python
new=df.drop(notlist,axis=0)
final=df.drop(removeList,axis=0)
new['budget']=new.budget.astype(float)
final['budget'] = final.budget.astype(float)
print(new['budget'])
print(final['budget'])
```
